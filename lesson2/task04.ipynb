```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
matplotlib.pyplot.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('creditcard.csv')
data.head()
```

```python
class_c = data["Class"].value_counts()
class_c
```

```python
class_c.plot(kind="bar")
plt.show()
```

```python
class_c.plot(kind="bar")
plt.yscale(value="log")
plt.show()
```

```python
gist_1 = data[data["Class"] > 0].V1
gist_1
```

```python
gist_0 = data[data["Class"] == 0].V1
gist_0
```

```python
plt.hist(gist_1, bins=20, density=True, color="#ff4c4c", alpha=0.5, label="Class 1")
plt.hist(gist_0, bins=20, density=True, color="grey", alpha=0.5, label="Class 0")
plt.xlabel('V1')
plt.legend()
plt.show()
```
